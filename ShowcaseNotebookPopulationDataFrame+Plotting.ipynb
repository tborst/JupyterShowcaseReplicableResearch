```python
# import pandas as pd
import pandas as pd
```

```python
# Create an object of type:dataframe
# Read the .csv-data from the local file (here, the file is located in a
# subdirectory). Tabular data is sorted according to the continent followed by a
# continent's country
dataWorld = pd.read_csv('/content/gapminder_all.csv', index_col='country')
# Display the data with its first ten rows
dataWorld.head(10)
```

```python
# Let us explore the properties of the table data
# First, we would like to know more about the independent variables used in the table. What variables are used?
dataWorld.columns
# What is the type and range of values of the variables?
dataWorld.describe()
# Some general info on the table data?
dataWorld.info()
```

```python
# For better readability, turn all columns for gdpPercap, lifeExp and pop into
# non-floating integers
# First, create a new table dataPopLife by copying the table dataWorld
dataPopLife = dataWorld.copy()
# Then, delete all columns starting with 'gdpPercap' from the table 'dataPopLife'
cols_to_drop = [col for col in dataPopLife.columns if col.startswith('gdpPercap')]
dataPopLife.drop(cols_to_drop, axis=1, inplace=True)
# Then, turn all columns for lifeExp and pop into non-floating integers
dataPopLife[dataPopLife.columns[1:]] = dataPopLife[dataPopLife.columns[1:]].astype(int)
# Display first five rows of dataPopLife
dataPopLife.head(5)


```

```python

# What is now the type and range of values of the variables?
dataPopLife.info()
```

```python
# Plot the development of life expectancy for three random countries per continent by means of matplotlib library.
# If there are less than three countries, use any country from that continent

!pip install --upgrade pandas
import matplotlib.pyplot as plt
import random

# Select life expectancy columns
life_exp_cols = [col for col in dataPopLife.columns if col.startswith('lifeExp')]

# Extract years from column names
years = [int(col.split('_')[1]) for col in life_exp_cols]

# Group data by continent and select up to three random countries from each
# If a continent has fewer than three countries, it will sample all of them
sampled_data = dataPopLife.groupby('continent').apply(lambda x: x.sample(n=min(3, len(x)), random_state=42))

# Reset index for easier plotting, but drop the existing 'continent' index
# to avoid duplicate column names when it's re-inserted
sampled_data = sampled_data.reset_index(level=0, drop=True)  # Drop the 'continent' index from levels

# Now reset the index again to get a numerical index and keep 'continent' as a column
sampled_data = sampled_data.reset_index()

# Rename the 'index' column to 'continent_group' for clarity
sampled_data = sampled_data.rename(columns={'index': 'continent_group'})

# Plot only life expectancy for selected countries
fig, ax = plt.subplots(figsize=(10, 6))

for continent, group in sampled_data.groupby('continent'):
    for index, row in group.iterrows():
        country_name = row['country']
        ax.plot(years, row[life_exp_cols], label=f"{country_name} ({continent})")

# Customize the plot
ax.set_title('Life Expectancy Development by Country (3 per Continent)')
ax.set_xlabel('Year')
ax.set_ylabel('Life Expectancy')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()

```

```python
# For better understanding and analysis, introduce interactive labels in the plot
!pip install plotly

import plotly.express as px

# Melt the DataFrame to long format
life_exp_cols = [col for col in sampled_data.columns if col.startswith('lifeExp')]
melted_data = sampled_data.melt(id_vars=['country', 'continent'],
                                value_vars=life_exp_cols,
                                var_name='year',
                                value_name='lifeExp')

# Extract year from the 'year' column
melted_data['year'] = melted_data['year'].str.split('_').str[1].astype(int)

# Create a new column combining country and continent for the legend labels
melted_data['country_continent'] = melted_data['country'] + ' (' + melted_data['continent'] + ')'

# Create the interactive plot using the melted data
fig = px.line(melted_data,
              x='year',
              y='lifeExp',
              color='country_continent',  # Use the new column for color mapping
              line_group='continent',
              hover_name='country',
              title='Life Expectancy Development by Country (randomly three per Continent)')

fig.update_layout(
    xaxis_title="Year",
    yaxis_title="Life Expectancy",
    legend_title="Country (Continent)"  # Update legend title
)

fig.show()
```
